```python

from abc import ABC, abstractmethod
# ABC: class.  # abstractmethod: decorator

# Interface : 

class A(ABC):
    
    @abstractmethod
    def say_hello(self):
        print("hello")
    
    def say_by(self):
        print("BY")

class B(A): 
    def say_hello(self):
        return super().say_hello()

obj = B()
obj.say_hello()
obj.say_by()
```

```python
"""

MYSQL - get_connection, query , close_connection
POSTGRES -  get_connection, query , close_connection

"""

from abc import ABC, abstractmethod

class SQLInterface(ABC):
    @abstractmethod
    def get_connection(self): pass

    @abstractmethod
    def query(self): pass

    @abstractmethod
    def close_connection(self): pass

class Mysql(SQLInterface):
    def get_connection(self): pass
    def query(self): pass
    def close_connection(self): pass


class Postgres(SQLInterface):
    def get_connection(self): pass
    def query(self): pass
    def close_connection(self): pass

# Postgres()
```

```python
def get_db_connection(dbtype):
    if dbtype == "psql":
        return Postgres()
    else:
        return Mysql()
    
obj = get_db_connection("mysql")
```
