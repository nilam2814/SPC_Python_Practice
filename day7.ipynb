```python
""" Calling Function By Creating Object"""

class Math:
    def add(self, a,b):
        print("inside add function")
        return a+b
    
    def multiply(self, a, b):
        print("inside multiply")
        return a*b
    
# Create Object
obj = Math()
obj.multiply(2,4)
```

```python
""" Calling Function Inside Other Function of Class"""

class Math:
    def add(self, a,b):
        print("inside add function")
        return a+b
    
    def multiply(self, a, b):
        print("inside multiply")
        self.add(a, b)
        return a*b
    
# Create Object
obj = Math()
obj.multiply(2,4)
```

```python
""" Calling Function Inside Other Function of Class"""

class Math:
    def __init__(self, a, b): # constructor / optional
        print("Constructor called")
        self.a = a # instance variable
        self.b = b

    def add(self): # instance method (self--> instance)
        print("inside add function")
        return self.a +self.b
    
    def multiply(self):
        print("inside multiply")
        return self.a * self.b
    
# Create Object
a=5
b=20

obj = Math(a,b)
summ = obj.add()
mul = obj.multiply()
print(summ)
print(mul)
# mul = obj.multiply(a,b)
# obj.b
```

```python
# Instance Method
# Class Method 
# static method

# class variable
# instance variable

class Math:
    total = 0 # class variable


    def __init__(self):
        self.a=10  # instance vaiable 

    @staticmethod
    def say_hello():   # static method # instanc variable is not usable here
        print("hell")
        print(Math.total) # using class Name you can access class variable



Math.say_hello()

```

```python
class Math:
    total = 0 # class variable


    def __init__(self):
        self.a=10  # instance vaiable 

    @classmethod
    def say_hello(cls):   # static method # instanc variable is not usable here
        print("hell")
        print(cls.total) # using class Name you can access class variable



Math.say_hello()
```

```python
"""
add --> a+b == result 

result * 5

"""

def multiply_with_5(func): # add
    def wrapper(*args,**kwargs): # (a,b)  ,{}
        result = func(*args,**kwargs)
        mul = result * 5
        return mul
    return wrapper

@multiply_with_5
def add(a, b):
    return a+b

add(2,3)
```

```python
import time

def time_calculate(func):
    def inner(*args, **kwargs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time()
        print(f"Function {func.__name__} took time {end_time-start_time}")
        return None
    return inner

@time_calculate
def add():
    time.sleep(2)
    print("Function Complete add")

@time_calculate
def multiply():
    time.sleep(3)
    print("Function Complete Multiply")

add()
multiply()
```

```python
"""
a, b

Math 

add
multiply
devide

get_result
-- inp: ["add", "mul"]
-->Result : add, mu. dev --> {"add": mul, dev} didc


"""



class Math:
    def __init__(self,a,b):
        self.a=a
        self.b=b 

    def add(self):
        return self.a + self.b

    def multiply(self):
        return self.a * self.b

    def devide(self):
        return self.a % self.b

    def get_result(self, calc_list): 
        print(calc_list)
        result = {}
        for fun in calc_list:
            if fun =="add":
                addition = self.add()
                result[fun] = addition
            elif fun=="mul":
                multiplication= self.multiply()
                result[fun] = multiplication
            elif fun =="dev":
                dev_result = self.devide()
                result[fun]= dev_result
        return result




obj = Math(20,30)
obj.get_result(["add", "dev"])
```
