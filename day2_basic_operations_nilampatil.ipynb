### Set functions

```python
s={1,2,3}
s
```

```python
s.add(4)#Adds a single element to the set.
s
```

```python
s.update([5,6,7])#Adds multiple elements (iterable) to the set.
s
```

```python
s.remove(4)# Removes an element, shows KeyError if the element is not present.
print(s)
```

```python
s.discard(6) #Removes an element,Does NOT raise an error if the element is missing.
s
```

```python
x=s.pop() #Removes and returns a random element (since sets are unordered).
print(x)
print(s)
```

```python
s.clear() #Removes all elements from the set.
print(s)
```

```python
a={1,2,3}
b={3,4,5}

print(a)
print(b)
```

```python
print(a.union(b))   # Returns a new set containing all unique elements from both sets.
```

```python
print(a.intersection(b))   # Returns common elements from both sets.
```

```python
print(a.difference(b)) #Returns elements present in the first set but not in the second.
```

```python
print(a.symmetric_difference(b))#eturns elements that are in either set, but not in both.
```

```python
a={1,2}
b={1,2,3}
print(a.issubset(b))# Checks if all elements of one set exist in another.
```

```python
x={1,3,2}
y={2,3}
print(x)
print(x.issuperset(y)) #Checks if one set contains all elements of another. 
```

```python
print(a.isdisjoint(b))#Returns True if two sets have no common elements.
```

```python
a = {1, 2, 3, 2}
b= a.copy() # Returns a shallow copy of the set.
print(b)
```

### Dictionary operations

```python
#1.create dictionary 
student={"name":"Nilam", "age":25, "city":"Pune"}
```

```python
#2.Access dictionary value
print(student["name"]) # can access value using key
```

```python
print(student.get("age"))# can access value using get(), get does not throw error if key is missing thats why get() is safer to use
```

```python
#3. Add or update key value
student["email"]= "nilam@gmail.com"  # adding value to dictionary
student["age"]=26 # update value
student
```

```python
#4. Delete oprations
student.pop("age") # it removes item & returns value
print(student)
```

```python
student.popitem()# it removes last inserted item
print(student)
```

```python
del student["email"] #removes by key
print(student)
```

```python
student.clear() #removes all items
print(student)
```
