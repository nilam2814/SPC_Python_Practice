# Python OOP Assignment — Inheritance Examples

```python
# 1) Single Inheritance: Person -> Student
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no

# TODO:
# 1. Create object of Student.
# 2. Print name.
# 3. Print roll_no.

std=Student("nilam",101)
print(std.name)
print(std.roll_no)
```

```python
# 2) Single Inheritance: Animal -> Dog
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

# TODO:
# 1. Create Dog object.
# 2. Call sound() from parent class.
# 3. Call bark() from child class.

d=Dog()
d.sound()
d.bark()
```

```python
# 3) Constructor Inheritance: Vehicle -> Car
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)  
        self.model = model

# TODO:
# 1. Create Car object.
# 2. Print brand.
# 3. Print model.

c=Car("Tata", "Harrier")
print(c.brand)
print(c.model)
```

```python
# 4) Multilevel Inheritance: A -> B -> C
class A:
    def showA(self):
        print("This is A")

class B(A):
    def showB(self):
        print("This is B")

class C(B):
    def showC(self):
        print("This is C")

# TODO:
# 1. Create C object.
# 2. Call showA(), showB(), showC().

c=C()
c.showA()
c.showB()
c.showC()
```

```python
# 5) Multilevel Inheritance with methods: LivingThing -> Animal -> Dog
class LivingThing:
    def living(self):
        print("I am alive")

class Animal(LivingThing):
    def eat(self):
        print("Eating…")

class Dog(Animal):
    def bark(self):
        print("Barking…")

# TODO:
# 1. Create Dog object.
# 2. Call living(), eat(), bark().

d=Dog()
d.living()
d.eat()
d.bark()
```

```python
# 6) Multiple Inheritance: Father + Mother -> Child
class Father:
    def father_method(self):
        print("Father method")

class Mother:
    def mother_method(self):
        print("Mother method")

class Child(Father, Mother):
    def child_method(self):
        print("Child method")

# TODO:
# 1. Create Child object.
# 2. Call father_method(), mother_method(), child_method().
c=Child()
c.father_method()
c.mother_method()
c.child_method()
```

```python
# 7) Constructor Inheritance: Product -> Laptop
class Product:
    def __init__(self, name):
        self.name = name

class Laptop(Product):
    def __init__(self, name, ram):
        super().__init__(name)
        self.ram = ram

# TODO:
# 1. Create Laptop object.
l=Laptop("mobile", 8)
# 2. Print name.
print(l.name)
# 3. Print ram.
print(l.ram)
```

```python
# 8) Multilevel with __init__: Country -> State -> City
class Country:
    def __init__(self, cname):
        self.cname = cname

class State(Country):
    def __init__(self, cname, sname):
        super().__init__(cname)
        self.sname = sname

class City(State):
    def __init__(self, cname, sname, cityname):
        super().__init__(cname, sname)
        self.cityname = cityname

# TODO:
# 1. Create City object.
c=City("India", "Maharashtra", "Mumbai")
# 2. Print cname, sname, cityname.
print(c.cname)
print(c.sname)
print(c.cityname)
```

```python
# 9) Multiple Inheritance with methods: Teacher + Player -> Student
class Teacher:
    def teach(self):
        print("Teaching…")

class Player:
    def play(self):
        print("Playing…")

class Student(Teacher, Player):
    def study(self):
        print("Studying…")

# TODO:
# 1. Create Student object.
t=Student()

# 2. Call teach(), play(), study().
t.teach()
t.play()
t.study()
```

```python
# 10) Constructor Inheritance: Shape -> Rectangle
class Shape:
    def __init__(self, color):
        self.color = color

class Rectangle(Shape):
    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width

# TODO:
# 1. Create Rectangle object.
r=Rectangle("Red",4,5)

# 2. Print color, length, width.
print(r.color)
print(r.length)
print(r.width)
```
