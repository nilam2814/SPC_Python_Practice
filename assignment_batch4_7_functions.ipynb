### 14.Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
operations_list = [
    {"op": "add", "a": 10, "b": 20},
    {"op": "mul", "a": 4, "b": 5}
]
def adapter(operations):
    result=[]

    for item in operations:
        op=item["op"]
        a=item["a"]
        b=item["b"]

        if op=="add":
            result.append(a+b)
        elif op=="sub":
            result.append(a-b)
        elif op=="mul":
            result.append(a*b)
        elif op=="div":
            if b!=0:
                result.append(a/b)
            else:
                result.append("Division by zero")
        else:
            result.append("Invalid data")
    return result

adapter(operations_list)
```

### 15.Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
student_data= [ 
    {"name":"A", "marks":[80,90,70]}, 
    {"name":"B", "marks":[60,75,65]} 
] 
def process_students(data):
    
    for values in data:
        name=values["name"]
        marks=values["marks"]
        total=0
        grade=""
        for m in marks:
            total=total+m

        if total>=240:
            grade="A"
            
        elif total>=200:
            grade="B"
            
        else:
            grade="C"
            

        print(f"Name:{name}| Total:{total} |Grade:{grade}")

process_students(student_data)
```

### 16.Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
products_list=[ 
    {"name":"Laptop", "price":50000, "discount":10}, 
    {"name":"Phone", "price":20000, "discount":5} 
    ]

def apply_discounts(products):

    for items in products:
        name=items["name"]
        price=items["price"]
        discount=items["discount"]

        final_price= price-(price*discount/100)

        print(f"{name} final:{final_price}")
apply_discounts(products_list)

```
