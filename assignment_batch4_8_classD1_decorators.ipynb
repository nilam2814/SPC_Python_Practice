**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
# class Date:

#     @classmethod
#     def from_string(cls, s):

# class Date:
   
#     @classmethod
#     def from_string(cls, s):
#         day, month, year = s.split("-")
#         # return cls(int(day), int(month), int(year))
#         return day,month,year


# Date.from_string("28-06-2000")

class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, date_str):
        day, month, year = date_str.split("-")
        return cls(int(day), int(month), int(year))
d1 = Date.from_string("05-10-2025")
print(d1.day, d1.month, d1.year)      
```

## G. Decorators (27–30)


**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
# TODO: Write your solution for Q27 below

def decor(func):
    def inner():
        
        print("Function started")
        func()
        
    return inner

@decor
def abc():
    print("welcome")

abc()
```

**Q28.** Create a decorator that prints `"Function ended"` after running the function. Test it.

```python
# TODO: Write your solution for Q28 below

def decor(func):
    def inner():
        func()
        print("function ended")
    return inner
        

@decor
def abc():
    print("welcome")
    print("nilam")
    print("patil")

abc()
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
# TODO: Write your solution for Q29 below

def decor(func):
    def inner(*args,**kwargs):
        print("function name--->",func.__name__)
        return func(*args,**kwargs)
    return inner

@decor
def call_function():
    pass

call_function()
```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python
# TODO: Write your solution for Q30 below

def decor1(func):
    def inner():
        print("start")
        func()
    return inner

def decor2(func):
    def inner():
        func()
        print("end")
    return inner

@decor1
@decor2
def nilam():
    print("hello world")

nilam()
```
