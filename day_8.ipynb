```python
""" Inheritance 

- Single A--B
- Multilevel A-->B-->C
- Multiple
A. -->C
B. -->


"""
```

```python
# - Single A--B


class A:
    def method_a(self):
        print("Method A called")

class B(A):
    def method_b(self):
        print("Method B called")

obj_b = B()
# obj_b.method_b()
obj_b.method_a()
```

```python
# - Multilevel A--B--C


class A:
    def method_a(self):
        print("Method A called")

class B(A):
    def method_b(self):
        print("Method B called")

class C(B):
    def method_c(self):
        print("Method C called")

obj_c = C()
obj_c.method_c()
obj_c.method_b()
obj_c.method_a()
```

```python
# Multiple

class A:
    def method_a(self):
        print("Method A called")

class B:
    def method_b(self):
        print("Method B called")

class C(B, A):
    def method_c(self):
        print("Method C called")

obj_c = C()
obj_c.method_c()
obj_c.method_b()
obj_c.method_a()
```

```python
# - Single A--B


class A:
    def __init__(self, a, b):
        self.a = a 
        self.b = b

    def method_a(self):
        print("Method A called")

class B(A):
    def __init__(self, a, b):
        print("COnstructor B ")
        super().__init__(a, b)

    def method_b(self):
        print("Method B called")

obj_b = B(10, 20)
obj_b.a
```

```python
def check_permission(funcs):
    def inner(*args, **kwargs):
        print(kwargs)
        if  kwargs["username"] == "u" and  kwargs["password"]=="p":
            rsult = funcs(*args, kwargs)
            return rsult
        else:
            print("Invalid credentials")
            return None
    return inner

@check_permission
def api_call1(inp):
    return "data1"

@check_permission
def api_call2(inp):
    return "data2"
```
